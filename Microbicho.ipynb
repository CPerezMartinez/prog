```python
import pandas as pd

df=pd.read_csv("microbicho - Full 1 (3).csv")
df
```

```python
# @title Temps seg vs Intensitat de la llum a14

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Temps seg']
  ys = series['Intensitat de la llum a14']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Temps seg', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Temps seg')
_ = plt.ylabel('Intensitat de la llum a14')
```
